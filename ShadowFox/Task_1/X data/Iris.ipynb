```python
import numpy as np
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
sns.set()
```

```python
raw_data = pd.read_csv("/content/iris.data.csv")
raw_data
```

```python
data = raw_data.copy()
data.describe()
```

```python
data['Iris-setosa'] = data['Iris-setosa'].map({'Iris-setosa':1,'Iris-versicolor':2,'Iris-virginica':3})
X=data.iloc[:,4:5].values
X
```

```python

kmeans = KMeans(3)
kmeans.fit(X)
```

```python
identified_clusters = kmeans.fit_predict(X)
identified_clusters
```

```python
plt.scatter(data['5.1'] ,data['3.5'] ,c = identified_clusters ,cmap = 'rainbow')
```

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(data)
x_scaled
```

```python
kmeans1 = KMeans(2)
kmeans1.fit(x_scaled)
```

```python
identified_clusters1 = kmeans1.fit_predict(x_scaled)
identified_clusters1
```

```python
plt.scatter(data['5.1'] ,data['3.5'] ,c = identified_clusters1 ,cmap = 'rainbow')
```

```python
wcss = []
col_num = 10
for i in range(1,col_num):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

wcss
```

```python
number_clusters = range(1,col_num)
plt.plot(number_clusters, wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```
